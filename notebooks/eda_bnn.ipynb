```python
from self_interferometry.acquisition.simulations.interferometers import (
    MichelsonInterferometer,
)
from self_interferometry.acquisition.simulations.coil_driver import CoilDriver
from self_interferometry.acquisition.simulations.waveform import Waveform
import numpy as np
import matplotlib.pyplot as plt
```

```python
interferometer_515 = MichelsonInterferometer(wavelength=0.515, phase=0)
```

```python
waveform_generator = Waveform(
    start_freq=10,
    end_freq=1000,
    gen_dec=8192,
    acq_dec=256,
    allowed_freqs=None,  # Use all valid frequencies in the range
)

coil_driver = CoilDriver()
time, voltage, voltage_spectrum = coil_driver.sample(
    waveform_generator, normalize_gain=False
)
plt.plot(time, voltage)
```

```python
displacement = coil_driver.get_displacement(voltage, sample_rate=125e6)
displacement = displacement[0] * 1e6  # convert to microns
plt.plot(time, displacement)
```

```python
time, signals, displacement, velocity = interferometer_515.get_simulated_buffer(
    displacement=displacement, time=time
)
```

```python
plt.plot(time[0:512], signals[0:512])
plt.plot(time[0:512], (displacement - displacement[0])[0:512])
```
